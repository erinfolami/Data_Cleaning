```
import pandas as pd
```

```
df = pd.read_csv("sentiment_transformers.csv")
df = df.iloc[:,1:]
df["label"] = df["sentiment"].str.extract(r'([A-Z]\w+ *)')
df["score"] = df["sentiment"].str.extract(r'(\d.\d+ *)')
df = df[["label","score"]]
df = df.to_csv("data_cleaned.csv",index=None)
print(df)
```
